```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import h5py
from scipy import stats
from scipy.io import loadmat
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression, LogisticRegression
from scipy.interpolate import interp1d
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import KFold, StratifiedKFold
from sklearn.datasets import load_iris
import statsmodels.stats.api as smstats
from scipy.stats import wilcoxon
```

```python
pathname_read = r'/Users/ygao/Desktop/Yifan/spring_2025/FNDIV/PTM'

if os.path.exists(pathname_read):
    print("The directory exists.")
else:
    print("The directory does not exist.")

subjs = ['0020', '0021', '0022', '0023', '0024', '0025', '0026', '0027', '0028', '0029',
        '0030', '0031', '0032', '0033', '0034', '0035', '0036', '0037', '0038',
        '0040', '0041', '0042', '0043', '0044', '0045', '0046', '0047', '0048', '0049',
        '0050', '0051', '0052', '0053', '0054', '0055', '0057', '0058', '0059',
        '0060', '0061', '0062', '0063', '0064', '0065', '0066', '0067', '0068', '0069']
#0056 didn't have EEG
#0039 had insufficient segments for k-fold cross-validation
n_subjs = len(subjs)
```

```python
sampling_rate = 0.2 #(kHz)
if sampling_rate == 1:
    acc_early = pd.read_csv('accuracy_early_all.csv')
    acc_late = pd.read_csv('accuracy_late_all.csv')
    time_points = np.linspace(-200, 1500, 1700)
    plot_name = 'accuracy_1000Hz.jpg'
elif sampling_rate == 0.2:
    acc_early = pd.read_csv('accuracy_early_all_200Hz.csv')
    acc_late = pd.read_csv('accuracy_late_all_200Hz.csv')
    time_points = np.linspace(-200, 1500, 340)
    plot_name = 'accuracy_200Hz.jpg'
```

## Optional - plot all subjects

```python
# all subjects - early
plt.figure(figsize=(14, 6)) 
for i in range(min(n_subjs, acc_early.shape[0])): 
    plt.plot(time_points, acc_early.iloc[i], label=f"Subject {subjs[i]}") 
plt.title("Early Block Accuracy for Each Subject")
plt.xticks(np.arange(-200, 1501, 100))
plt.xlabel('Time[ms] (0 = cue onset)')
plt.ylabel('Average Decoding Accuracy')
plt.show()

# all subjects - late
plt.figure(figsize=(14, 6))
for i in range(min(n_subjs, acc_late.shape[0])):  # Loop through each subject
    plt.plot(time_points, acc_late.iloc[i], label=f"Subject {subjs[i]}")  # Plot accuracy for each subject
plt.title("Late Block Accuracy for Each Subject")
plt.xticks(np.arange(-200, 1501, 100))
plt.xlabel('Time[ms] (0 = cue onset)')
plt.ylabel('Average Decoding Accuracy')
plt.show()
```

# T-tests across time and plot

```python
# run t-tests
t_stat, p_values = stats.ttest_rel(acc_early, acc_late, axis=0)
reject, pvals_corrected = smstats.multipletests(p_values, alpha=0.05, method='fdr_bh')[:2]
sig_timepoints_fdr = time_points[reject]  # Indices where p-values are below 0.05 after FDR correction
sig_timepoints = time_points[p_values < 0.05]  # Time points with p-value < 0.05

print(t_stat.shape)
print(p_values.shape)

# compute the average early and late accuracies across time for plotting
avg_acc_early = np.mean(acc_early, axis=0)
avg_acc_late = np.mean(acc_late, axis=0)

plt.figure(figsize=(14, 6))
plt.plot(time_points, avg_acc_early, color='palevioletred', label='Early Block')
plt.plot(time_points, avg_acc_late, color='slateblue', label='Late Block')
plt.scatter(sig_timepoints_fdr, np.ones_like(sig_timepoints_fdr)*min(avg_acc_early), color='red', label='p < 0.05 FDR corrected')
plt.scatter(sig_timepoints, np.ones_like(sig_timepoints)*min(avg_acc_early), color='orange', label='p < 0.05')

plt.xticks(np.arange(-200, 1501, 100))
plt.xlabel('Time[ms] (0 = cue onset)')
plt.ylabel('Average Decoding Accuracy')
plt.title('Average Accuracy Across Time Points (Early vs Late Blocks)')
plt.legend()
plt.show()
plt.savefig(plot_name)
```
